```python
# shroom.drop(['class', 'population', 'habitat'], axis=1, inplace=True)

index = pd.MultiIndex.from_frame(shroom[['class', 'population', 'habitat']])
```


6. Use the .loc[] notation you learned and list the data about the edible
mushrooms.

```python
shroom.drop(['class', 'population', 'habitat'], axis=1, inplace=True)

shroom.set_index(index, inplace = True)
```
